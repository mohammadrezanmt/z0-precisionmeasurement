# **Exercise 1 : Optimize lepton Selection**

Firstly, the regarding libraries are imported as follow


```python
import uproot
import awkward as ak
import mplhep
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import math
```

Then, the path of the monte carlo(simulations for 4 channels) files are defined and imported 
them for 4 different variables

```python
##Defining the path
path_mc = '/home/mohammadreza/mc_z0experiment/'

##importing the data and different variables

ee = uproot.open(path_mc +'ee.root')
mm = uproot.open(path_mc +'mm.root')
tt = uproot.open(path_mc +'tt.root')
qq = uproot.open(path_mc +'qq.root')

ttree = "myTTree"

##print list of branches
ee[ttree].keys()
mm[ttree].keys()
tt[ttree].keys()
qq[ttree].keys()

##load branches
branches_ee = ee[ttree].arrays()
branches_mm = mm[ttree].arrays()
branches_tt = tt[ttree].arrays()
branches_qq = qq[ttree].arrays()

##define a numpy array for the different channels and variavles

varp = 'Pcharged'
varn = 'Ncharged'
vare = 'E_ecal'
varh = 'E_hcal'

###Pcharged

pchar_ee = ak.to_numpy(branches_ee[varp])
pchar_mm = ak.to_numpy(branches_mm[varp])
pchar_tt = ak.to_numpy(branches_tt[varp])
pchar_qq = ak.to_numpy(branches_qq[varp])

###Ncharged

nchar_ee = ak.to_numpy(branches_ee[varn])
nchar_mm = ak.to_numpy(branches_mm[varn])
nchar_tt = ak.to_numpy(branches_tt[varn])
nchar_qq = ak.to_numpy(branches_qq[varn])

###ECAL

ecal_ee = ak.to_numpy(branches_ee[vare])
ecal_mm = ak.to_numpy(branches_mm[vare])
ecal_tt = ak.to_numpy(branches_tt[vare])
ecal_qq = ak.to_numpy(branches_qq[vare])

###HCAL

hcal_ee = ak.to_numpy(branches_ee[varh])
hcal_mm = ak.to_numpy(branches_mm[varh])
hcal_tt = ak.to_numpy(branches_tt[varh])
hcal_qq = ak.to_numpy(branches_qq[varh])
```

Now, we can plot the different channel with different variables and we can estimate a sensible range for each variables

```python
##plotting  Pcharged 

plt.style.use(mplhep.style.ATLAS)
plt.figure(figsize=(7,5))

pe_c, pe_e,_ = plt.hist(pchar_ee,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='b', hatch='/', label='electron')
pm_c, pm_e,_ = plt.hist(pchar_mm,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='r', hatch='/', label='muon') 
pt_c, pt_e,_ = plt.hist(pchar_tt,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='g', hatch='/', label='tau') 
pq_c, pq_e,_ = plt.hist(pchar_qq,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='y', hatch='/', label='hadron')

plt.legend(loc='upper left')
plt.title('Pcharged')
plt.xlim(10.,105.)
plt.ylim(0.,1500.)
plt.xlabel('Total scalar sum of track momenta')
plt.ylabel('Number of events')
plt.show()


##plotting  Ncharged 

plt.style.use(mplhep.style.ATLAS)
plt.figure(figsize=(7,5))


binn = np.linspace(0,1000,1000)
ne_c, ne_e,_ = plt.hist(nchar_ee,bins=binn,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='b', hatch='/', label='electron' ) 
nm_c, nm_e,_ = plt.hist(nchar_mm,bins=binn,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='r', hatch='/', label='muon')
nt_c, nt_e,_ = plt.hist(nchar_tt,bins=binn,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='g', hatch='/', label='tau') 
nq_c, nq_e,_ = plt.hist(nchar_qq,bins=binn,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='y', hatch='/', label='hadron')

plt.legend(loc='upper right')
plt.title('Ncharged')
plt.xlim(0.,35.)
plt.ylim(0.,100000.)
plt.xlabel('Total scalar sum of track momenta')
plt.ylabel('Number of events')
plt.show()


##plotting  ECAL 

plt.style.use(mplhep.style.ATLAS)
plt.figure(figsize=(7,5))

ee_c, ee_e,_ = plt.hist(ecal_ee,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='b', hatch='/', label='electron')
em_c, em_e,_ = plt.hist(ecal_mm,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='r', hatch='/', label='muon')
et_c, et_e,_ = plt.hist(ecal_tt,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='g', hatch='/', label='tau')
eq_c, eq_e,_ = plt.hist(ecal_qq,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='y', hatch='/', label='hadron')

#l = np.sqrt(h)

#m = np.sum(h)

plt.legend(loc='upper right')
plt.title('ECAL')
plt.xlim(0.,105.)
plt.ylim(0.,3500.)
plt.xlabel('Total energy measured in the electromagnetic calorimeter, GeV')
plt.ylabel('Number of events')
plt.show()

#Plotting HCAL
plt.style.use(mplhep.style.ATLAS)
plt.figure(figsize=(7,5))

he_c, he_e,_ = plt.hist(hcal_ee,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='b', hatch='/', label='electron') 
hm_c, hm_e,_ = plt.hist(hcal_mm,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='r', hatch='/', label='muon')
ht_c, ht_e,_ = plt.hist(hcal_tt,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='g', hatch='/', label='tau')
hq_c, hq_e,_ = plt.hist(hcal_qq,bins=1000,range=(0.,100.), histtype='step',  linewidth=2, edgecolor='y', hatch='/', label='hadron')

plt.legend(loc='upper right')
plt.title('HCAL')
plt.xlim(0.,50.)
plt.ylim(0.,3500.)
plt.xlabel('Total energy measured in the hadronic calorimete, GeV')
plt.ylabel('Number of events')
plt.show()


#for example
print(f"For example for number of events in Ncharged detection")
print(ne_c)
print(np.sum(ne_c))
print(len(ne_c))
```

```python
#Selection


#defining some new variables for all the cuts over MC simulations

msk_pe = pchar_ee <= 100
msk_pm = pchar_mm <= 100
msk_pt = pchar_tt <= 60
msk_pq = pchar_qq <= 80

msk_ne = nchar_ee <= 5
msk_nm = nchar_mm <= 4
msk_nt = nchar_tt <= 6
msk_nq = nchar_qq <= 30

msk_ee = ecal_ee <= 100 
msk_em = ecal_mm <= 20
msk_et = ecal_tt <= 60
msk_eq = ecal_qq <= 80

msk_he = hcal_ee <= 10 
msk_hm = hcal_mm <= 19
msk_ht = hcal_tt <= 25
msk_hq = hcal_qq <= 30


#for example

print(f"For example for electron channel:")

print(f"A total of '{sum(msk_pe)}' in Pcharged for the electron channel out of '{len(msk_pe)}' events passed the cut 'mask'")

print(f"A total of '{sum(msk_ne)}' in Ncharged for the electron channel out of '{len(msk_ne)}' events passed the cut 'mask'")

print(f"A total of '{sum(msk_ee)}' in ECAL for the electron channel out of '{len(msk_ee)}' events passed the cut 'mask'")

print(f"A total of '{sum(msk_he)}' in HCAL for the electron channel out of '{len(msk_he)}' events passed the cut 'mask'")
```

Now, we can apply some cuts over different channels(ee,tt,mm,qq) for different variables on the Opal data set. In order to apply the cut, we can use the definition of the deviation and mean of the data set. In other words, the best fit would be obtained when the deviation is less enough in compare to the whole data set by chosing a propriate cut. For finding the suitable cut we need to have less deviation value and you can see an example as follow for ee-channel on ECAL data of Monte carlo simulation.

```python
## Before chosing a cut over ee-channel in ECAL

print(f"Mean of {vare}: ({ecal_ee.mean()}) {np.nanmean(ecal_ee)}")
print(f"Standard deviation of {vare}: ({ecal_ee.std()}) {np.nanstd(ecal_ee)}")
print(f"Minimum value of {vare}: ({ecal_ee.min()}) {np.nanmin(ecal_ee)}")
print(f"Maximum value of {vare}: ({pchar_ee.max()}) {np.nanmax(ecal_ee)}")
```

However, after chosing the best cut we can see the deviation tend to be less than what we had. 

```python
## After chosing a cut over ee-channel in ECAL

print(f"Mean of {vare}: ({ecal_ee[msk_ee].mean()}) {np.nanmean(ecal_ee[msk_ee])}")
print(f"Standard deviation of {vare}: ({ecal_ee[msk_ee].std()}) {np.nanstd(ecal_ee[msk_ee])}")
print(f"Minimum value of {vare}: ({ecal_ee[msk_ee].min()}) {np.nanmin(ecal_ee[msk_ee])}")
print(f"Maximum value of {vare}: ({ecal_ee[msk_ee].max()}) {np.nanmax(ecal_ee[msk_ee])}")
```

this cut in the MC simulation is not only can lead to find ee-channel in the real data but also it can reduce the uncertainity of the final fit to the data.

```python
#Making an uncertainity matrix

M_mc = np.array([[sum(msk_pe)/len(msk_pe),sum(msk_pm)/len(msk_pm),sum(msk_pt)/len(msk_pt),sum(msk_pq)/len(msk_pq)],
               [sum(msk_ne)/len(msk_ne),sum(msk_nm)/len(msk_nm),sum(msk_nt)/len(msk_nt),sum(msk_nq)/len(msk_nq)],
               [sum(msk_ee)/len(msk_ee),sum(msk_em)/len(msk_em),sum(msk_et)/len(msk_et),sum(msk_eq)/len(msk_eq)],
               [sum(msk_he)/len(msk_he),sum(msk_hm)/len(msk_hm),sum(msk_ht)/len(msk_ht),sum(msk_hq)/len(msk_hq)]])

print(M_mc)
```
